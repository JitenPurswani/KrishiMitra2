```python
import pandas as pd
import joblib
import requests
import geocoder
import datetime
import os
from sklearn.preprocessing import LabelEncoder

# Load trained model & encoders safely
def safe_load(filename, default=None):
    if os.path.exists(filename):
        return joblib.load(filename)
    print(f"Warning: {filename} not found.")
    return default

model = safe_load("crop_recommendation_model_v2.pkl")
label_encoders = safe_load("label_encoders_v2.pkl", {})

# Load taluka-to-district mapping
try:
    mapping_df = pd.read_csv("maharashtra_taluka_district.csv")
except FileNotFoundError:
    print("Warning: District mapping file not found.")
    mapping_df = pd.DataFrame(columns=["District", "Taluka"])

trained_districts = set(label_encoders.get("location", {}).classes_) if hasattr(label_encoders.get("location", {}), "classes_") else set()

# Important: Get the EXACT feature names and order from the model
# This is crucial for scikit-learn models
if model is not None and hasattr(model, 'feature_names_in_'):
    FEATURE_NAMES = model.feature_names_in_.tolist()
else:
    # Fallback to what we think was used during training
    FEATURE_NAMES = ["N", "P", "K", "ph", "temperature", "humidity", "rainfall", "location", "season"]

# Hard-coded crop mapping - use this regardless of label encoders
CROP_MAPPING = {
    0: "Apple", 
    1: "Banana", 
    2: "Blackgram", 
    3: "Chickpea", 
    4: "Coconut",
    5: "Coffee", 
    6: "Cotton", 
    7: "Grapes", 
    8: "Jute", 
    9: "Kidney Beans",
    10: "Lentil", 
    11: "Maize", 
    12: "Mango", 
    13: "Moth Beans", 
    14: "Mung Bean",
    15: "Muskmelon", 
    16: "Orange", 
    17: "Papaya", 
    18: "Pigeon Peas",
    19: "Pomegranate", 
    20: "Rice", 
    21: "Watermelon"
}

def get_user_location():
    """Get user's taluka (or city) using IP-based geolocation."""
    try:
        g = geocoder.ip("me")
        return g.city if g.ok else None
    except Exception as e:
        print(f"Error detecting location: {e}")
        return None

def map_taluka_to_district(taluka):
    """Map a taluka to its corresponding district."""
    if taluka is None:
        return None
    match = mapping_df[mapping_df["Taluka"].str.lower() == taluka.lower()]
    if not match.empty:
        district = match.iloc[0]["District"]
        return district if district in trained_districts else None
    return None

def get_weather_data(location):
    """Fetch real-time weather data for a given location."""
    API_KEY = "77dccca726aab222672f70fdc5320ba3"
    try:
        url = f"http://api.openweathermap.org/data/2.5/weather?q={location},IN&appid={API_KEY}&units=metric"
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            data = response.json()
            return {
                "temperature": data["main"]["temp"],
                "humidity": data["main"]["humidity"],
                "rainfall": data.get("rain", {}).get("1h", 0)
            }
    except Exception as e:
        print(f"Weather request failed: {e}")
    return {"temperature": 28, "humidity": 60, "rainfall": 0}  # Fallback values

def get_season():
    """Determine the current season based on the month."""
    month = datetime.datetime.now().month
    return "Monsoon" if month in [6, 7, 8, 9] else "Summer" if month in [3, 4, 5] else "Winter"

def fetch_soil_data(district, season):
    """Retrieve soil parameters based on location and season."""
    # TODO: Replace with real database query if available
    return {"N": 50, "P": 30, "K": 20, "ph": 6.5}

def predict_crops(input_data):
    """Predict the top 3 crops based on input parameters."""
    if model is None:
        return ["Rice", "Wheat", "Cotton"]
    try:
        # Create DataFrame with EXACTLY the same feature names and order as training
        input_df = pd.DataFrame([input_data], columns=FEATURE_NAMES)
        
        # Print the feature names for debugging
        print(f"Using features: {list(input_df.columns)}")
        
        # Make the prediction
        crop_probabilities = model.predict_proba(input_df)[0]
        top_3_indices = crop_probabilities.argsort()[-3:][::-1]
        
        # Convert indices to integers and use the hard-coded mapping
        indices_as_int = [int(idx) for idx in top_3_indices]
        print(f"Top 3 indices: {indices_as_int}")  # Debug
        
        return [CROP_MAPPING.get(idx, f"Unknown-{idx}") for idx in indices_as_int]
    
    except Exception as e:
        print(f"Prediction error: {e}")
        import traceback
        traceback.print_exc()  # More detailed error information
        return ["Rice", "Wheat", "Cotton"]

def main():
    """Main function to run the crop recommendation system."""
    print("=== Crop Recommendation System ===")
    
    user_taluka = get_user_location()
    user_district = map_taluka_to_district(user_taluka)
    
    if not user_district:
        print(f"Using default district as fallback.")
        user_district = list(trained_districts)[0] if trained_districts else "Nagpur"
    
    print(f"Using district: {user_district}")
    weather_data = get_weather_data(user_district)
    season = get_season()
    soil_data = fetch_soil_data(user_district, season)
    
    # Start with empty dictionary
    input_data = {}
    
    # Add all numeric features
    input_data.update(soil_data)
    input_data.update(weather_data)
    
    # Process location - encode it if encoder exists
    try:
        if "location" in label_encoders and hasattr(label_encoders["location"], "classes_") and user_district in label_encoders["location"].classes_:
            input_data["location"] = label_encoders["location"].transform([user_district])[0]
        else:
            print(f"Warning: District '{user_district}' not in trained districts. Using default value.")
            input_data["location"] = 0
    except Exception as e:
        print(f"Error encoding location: {e}")
        input_data["location"] = 0
    
    # Process season - encode it if encoder exists
    try:
        if "season" in label_encoders and hasattr(label_encoders["season"], "classes_") and season in label_encoders["season"].classes_:
            input_data["season"] = label_encoders["season"].transform([season])[0]
        else:
            # Fallback mapping
            input_data["season"] = {"Monsoon": 0, "Summer": 1, "Winter": 2}.get(season, 0)
    except Exception as e:
        print(f"Error encoding season: {e}")
        input_data["season"] = 0

    # Make sure to include all expected features with correct types
    for feature in FEATURE_NAMES:
        if feature not in input_data:
            print(f"Warning: Feature '{feature}' missing. Using default value 0.")
            input_data[feature] = 0
    
    # Debug the input data
    print("\nInput Data:")
    for key, value in input_data.items():
        print(f"{key}: {value}")
        
    recommended_crops = predict_crops(input_data)
    
    print("\n=== Results ===")
    print(f"Detected Location: {user_district}")
    print(f"Temperature: {weather_data['temperature']}°C")
    print(f"Humidity: {weather_data['humidity']}%")
    print(f"Rainfall: {weather_data['rainfall']} mm")
    print(f"Season: {season}")
    print(f"Top 3 Recommended Crops: {', '.join(recommended_crops)}")

if __name__ == "__main__":
    main()
```
